```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name='2023', skiprows=3)
```

```python
df.info()
```

```python
valores_nulos = df.isnull().sum()
valores_nulos
```

## Procesamiento para eliminar valores nulos en dataframe

```python
df['Folio'] = df['Folio'].fillna(method='bfill')
```

```python
df['TIPO GASTO'] = df['TIPO GASTO'].fillna(method='ffill')
```

```python
df['MP'] = df['MP'].fillna('P')
```

```python
poliza = list(df['Poliza'].mode())
```

```python
df['Poliza'] = df['Poliza'].fillna(poliza[0])
```

```python
df['FP'] = df['FP'].fillna(round(df['FP'].mean(),1))
```

```python
df['Otros'] = df['Otros'].fillna(0.0)
```

```python
valores_nulos = df.isnull().sum()
valores_nulos
```

```python
fig = plt.figure(figsize = (15, 8))
df.plot(kind='box', vert = False)
plt.title('Valores Atípicos del DataFrame')
plt.show()
```

## Procedimiento 'Desviación Estándar' para eliminar outliers en DataFrame

```python
y = df.drop(df.columns[[0,1,2,3,4,5,6,7,17,18,19]], axis=1)
Limite_Superior = y.mean() + 3 * y.std()
Limite_Inferior = y.mean() - 3 * y.std()
print('Límite superior permitido', Limite_Superior)
print('Límite inferior permitido', Limite_Inferior)
```

```python
df1 = df[(y<=Limite_Superior)&(y>=Limite_Inferior)]
df1
```

```python
df2 = df1.drop(df1.columns[[0,1,2,3,4,5,6,7,17,18,19]], axis = 1)
df2
```

```python
valores_nulos = df2.isnull().sum()
valores_nulos
```

```python
data_limpio = df2.copy()
data_limpio = data_limpio.fillna(round(df2.mean(), 1))
data_limpio
```

```python
valores_nulos = data_limpio.isnull().sum()
valores_nulos
```

```python
columnas = df[df.columns[[0,1,2,3,4,5,6,7,17,18,19]]]
Datoslimpios_std = pd.concat([columnas, data_limpio], axis = 1)
Datoslimpios_std= Datoslimpios_std.reindex(columns=[
    'Fecha','Folio','UUID','RFC','Proveedor',
    'TIPO GASTO','Descripción','MP','FP','TC','Importe',
    'IVA','RET ISR','RET IVA','Otros','TOTAL MX',
    'TOTAL SAT','Tipo','Status','Poliza'])
Datoslimpios_std
```

```python
fig = plt.figure(figsize = (15, 8))
Datoslimpios_std.plot(kind='box', vert = False)
plt.title('Valores Atípicos del DataFrame')
plt.show()
```

```python
Datoslimpios_std.to_csv('DF 2020 1, Desviacion Estandar.csv')
```

## Procedimiento 'Cuantiles' para sustituir outliers en DataFrame

```python
y = df2
percentile25 = y.quantile(0.25) #Q1
percentile75 = y.quantile(0.75) #Q3
iqr = percentile75 - percentile25

Limite_Superior_iqr = percentile75 + 1.5 * iqr
Limite_Inferior_iqr = percentile25 - 1.5 * iqr
print('Límite superior permitido', Limite_Superior_iqr)
print('Límite inferior permitido', Limite_Inferior_iqr)
```

```python
data3_iqr = df1[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
data4_iqr = data3_iqr.drop(df1.columns[[0,1,2,3,4,5,6,7,17,18,19]], axis = 1)
data4_iqr
```

```python
valores_nulos = data4_iqr.isnull().sum()
valores_nulos
```

```python
data5_iqr = data4_iqr.copy()
data5_iqr = data5_iqr.fillna(round(data4_iqr.mean(), 1))
data5_iqr
```

```python
valores_nulos = data5_iqr.isnull().sum()
valores_nulos
```

```python
columnas = df[df.columns[[0,1,2,3,4,5,6,7,17,18,19]]]
Datoslimpios_cua = pd.concat([columnas, data5_iqr], axis = 1)
Datoslimpios_cua= Datoslimpios_cua.reindex(columns=[
    'Fecha','Folio','UUID','RFC','Proveedor',
    'TIPO GASTO','Descripción','MP','FP','TC','Importe',
    'IVA','RET ISR','RET IVA','Otros','TOTAL MX',
    'TOTAL SAT','Tipo','Status','Poliza'])
Datoslimpios_cua
```

```python
fig = plt.figure(figsize = (15, 8))
Datoslimpios_cua.plot(kind='box', vert = False)
plt.title('Valores Atípicos del DataFrame')
plt.show()
```

```python
Datoslimpios_cua.to_csv('DF 2023 2, Cuartiles.csv')
```
